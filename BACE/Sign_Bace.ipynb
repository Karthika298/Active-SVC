```python
def VarianceRemoval(data,thresh):
    traindata = data
    selector = VarianceThreshold(thresh)
    selector.fit(traindata)
    data_var_free=traindata[traindata.columns[selector.get_support(indices=True)]]
    return data_var_free
def correlation_check(traindata,thresh):
    traindata = traindata    
    corr_matrix = traindata.corr()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
    to_drop = [column for column in upper.columns if any(upper[column] >thresh)]
    trainset=traindata.drop(traindata[to_drop], axis=1)
    return trainset
def handlemissingvalue(X_train):
    traindata = X_train
    X_train=X_train.replace([np.inf, -np.inf,""," "], np.nan)
    X_train=X_train.replace([""," "],np.nan)
    X_train.fillna(X_train.mean(), inplace=True)
    return X_train
from sklearn import preprocessing
def normalize(traindata):
    traindata = traindata
    try:
        x = traindata.values #returns a numpy array
    except:
        x=traindata
    min_max_scaler = preprocessing.MinMaxScaler()
    min_max_scaler.fit(x)
    x_scaled = min_max_scaler.transform(x)
    train_normal = pd.DataFrame(x_scaled)
    train_normal.columns =list(traindata.columns.values)
    return train_normal
def Process_data(traindata):
    traindata = traindata
    traindata=handlemissingvalue(traindata)
    traindata=normalize(traindata)
    traindata=VarianceRemoval(traindata,0.0)
    traindata=correlation_check(traindata,0.95)
    return traindata
final_data=Process_data(data)
```
